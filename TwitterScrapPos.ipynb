```python
import snscrape.modules.twitter as scrapTwitter
```

```python
import datetime
```

```python
dataFinal = datetime.date.today()
dataInicial = '2022-06-26'
```

```python
listaTweetsBrasilParalelo = []
```

```python
for i, tweet in enumerate(scrapTwitter.TwitterSearchScraper(f'{"Brasil Paralelo"} + since:{dataInicial} until:{dataFinal}').get_items()):
  if i > 500:
    break
  listaTweetsBrasilParalelo.append([tweet.date, tweet.id, tweet.content, tweet.username])
```

```python
listaTweetsBrasilParalelo
```

```python
import pandas as pd
dataFrameBrasilParalelo = pd.DataFrame(listaTweetsBrasilParalelo, columns=["Data", "Id", "Tweet", "Usuario"])
```

```python
dataFrameBrasilParalelo
```

```python
dataFrameBrasilParalelo.to_csv("BrasilParalelo.csv", sep="#")
```
